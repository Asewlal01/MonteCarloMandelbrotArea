```python
import matplotlib.pyplot as plt
from MonteCarloIntegration import *
```

```python
low = -2
high = 2
samples = 10**3

pureRandom = []
latinHypercube = []
orthogonal = []
iterations = []
```

```python
for iter in range(1, 50):
    iterations.append(iter)
    # Compute area
    A1 = randomSampling(low, high, iter, samples)
    pureRandom.append(A1)

    A2 = latinHypercubeSampling(low, high, iter, samples)
    latinHypercube.append(A2)

    A3 = orthogonalSampling(low, high, iter, samples)
    orthogonal.append(A3)
```

```python
plt.plot(iterations, pureRandom, label='Pure Random')
plt.plot(iterations, latinHypercube, label='Latin Hypercube')
plt.plot(iterations, orthogonal, label='Orthogonal')
plt.xlabel('Iterations')
plt.ylabel('Area')
plt.legend()
plt.grid()
plt.show()
```
